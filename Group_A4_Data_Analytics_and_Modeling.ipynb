````python
import warnings
import itertools
from statsmodels.tsa.statespace.sarimax import SARIMAX

# ... โค้ด Grid Search parameters ของคุณ ...

# --- เริ่มการซ่อนคำเตือน ---
warnings.filterwarnings("ignore")
````

````python
# =========================
# Auto ARIMA (SARIMA search) - concise
# =========================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pmdarima import auto_arima
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error

# 0) โหลด/เตรียม series (ปรับ path ถ้าจำเป็น)
df = pd.read_csv("/content/drive/MyDrive/Walmart/time_series.csv")
df['Date'] = pd.to_datetime(df['Date'])
series = df.groupby('Date')['Weekly_Sales'].sum().sort_index()
# resample ให้เป็นสัปดาห์ที่ตรงกับ data (เช่น 'W-FRI' ถ้าบันทึกทุกศุกร์)
series = series.resample('W-FRI').sum().interpolate(method='time')

# 1) แบ่ง train/test (time-based)
split = int(len(series) * 0.7)
train = series.iloc[:split]
test  = series.iloc[split:]

# 2) เรียก auto_arima (search เฉพาะ seasonal models, m=52)
# ปรับ start/max params ตามต้องการ (ไม่ให้ search กว้างเกินไปเพื่อความเร็ว)
auto_model = auto_arima(
    train,
    start_p=0, start_q=0,
    max_p=3, max_q=3,
    d=None,            # ให้ auto_arima เลือก d อัตโนมัติ
    seasonal=True,
    start_P=0, start_Q=0,
    max_P=2, max_Q=2,
    D=None,            # ให้ auto_arima เลือก D อัตโนมัติ
    m=52,              # seasonal period = 52 สัปดาห์
    trace=True,
    error_action='ignore',
    suppress_warnings=True,
    stepwise=True,
    n_jobs=1
)

print("\n=== Auto ARIMA summary ===")
print(auto_model.summary())

# 3) พยากรณ์ช่วง test (ใช้ model ที่ได้)
n_test = len(test)
pred_test = auto_model.predict(n_periods=n_test)
pred_index = test.index

# 4) Metrics
rmse = np.sqrt(mean_squared_error(test.values, pred_test))
mape = mean_absolute_percentage_error(test.values, pred_test) * 100
print(f"\nRMSE: {rmse:,.2f}")
print(f"MAPE: {mape:.2f}%")

# 5) Plot Actual vs Predicted
plt.figure(figsize=(12,5))
plt.plot(train.index, train.values, label='Train', color='steelblue')
plt.plot(test.index, test.values, label='Actual (Test)', color='orange')
plt.plot(pred_index, pred_test, label='Predicted (auto_arima)', color='green', linestyle='--')
plt.title('Auto ARIMA (SARIMA search) — Actual vs Predicted')
plt.xlabel('Date'); plt.ylabel('Weekly Sales')
plt.legend(); plt.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
````

````python
#import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
````

#### Partition data
เลือกเฉพาะ column ที่สนใจ

````python
df1 = df[['Store','Size','Weekly_Sales','CPI','Unemployment','Fuel_Price']]
df1
````

````python
# Calculate the mean of all columns in df1, grouped by Store
df2 = df1.groupby('Store').median()

# Display the result
display(df2.reset_index().head())
````

````python
plt.scatter(x = df2['Size'], y = df2['Weekly_Sales'])
plt.xlabel('Size')
plt.ylabel('Weekly_Sales')
````

````python
df2['Sales_Per_Size'] = df2['Weekly_Sales'] / df2['Size']
````

````python
df_store_median = df2.drop(columns = ['Size','Weekly_Sales'])
df_store_median.reset_index().head()
````

````python
features = [
    'CPI',
    'Unemployment',
    'Fuel_Price',
    'Sales_Per_Size'
    ]

X = df_store_median[features]
````

#### Normalize

````python
#ปรับ scale ข้อมูลให้อยู่ในหน่วยเดียวกันด้วยการทำ standard scaler
scaler = StandardScaler()
X_sd = scaler.fit_transform(X)
````

````python
X_sd
````

````python
# Plot histograms for each scaled feature
for i in range(X_sd.shape[1]):
    plt.figure(figsize=(8, 6)) # Create a new figure for each plot
    plt.hist(X_sd[:, i], bins=12)
    plt.title(f'Histogram of {features[i]} (Scaled)')
    plt.xlabel(features[i])
    plt.ylabel('Frequency')
    plt.show() # Display each plot immediately
````

#### Plot WCSS to decisision K

````python
# ทำการสร้างโมเดลและพล็อตค่า error ในแต่ละ k เพื่อดูว่า ควรเลือกใช้ K เป็นเท่าไหร่
#โดยดูที่ึการลดลงของ error ไม่ได้ลดลงอย่างค้มค่า หรือ ดูที่มุมหักศอก

inertia = []
K = range(1, 10)
for k in K:
    model = KMeans(n_clusters=k, random_state=42)
    model.fit(X_sd)
    inertia.append(model.inertia_)

plt.plot(K, inertia, 'o-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.title('Elbow Method for Optimal k with StandardScaler')
plt.show()
````

#### Model selection
k=*4*

````python
#ใช้ K = 4
#add cluster in rawdata
kmeans = KMeans(n_clusters=4, random_state=42)
df_store_median['Cluster'] = kmeans.fit_predict(X_sd)
````

#### Gruop by cluster

````python
# แบ่งกลุ่ม
# Include 'Size' in the features for calculating the mean
cluster_profile = df_store_median.groupby('Cluster')[['CPI', 'Unemployment', 'Fuel_Price', 'Sales_Per_Size']].median().round(2)
cluster_profile.reset_index()
````

````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import StandardScaler

# ข้อมูล
data = {
    'Cluster_sd': [0, 1, 2, 3],
    'CPI': [135.08, 215.11, 129.08, 135.08],
    'Unemployment': [7.77, 7.93, 13.74, 8.07],
    'Fuel_Price': [3.54, 3.29, 3.70, 3.69],
    'Sales_Per_Size': [0.08, 0.05, 0.05, 0.05]
}

cluster_profile = pd.DataFrame(data)

# เลือกคอลัมน์ที่จะทำ radar chart
categories = ['CPI', 'Unemployment', 'Fuel_Price', 'Sales_Per_Size']

# Standardize
scaler = StandardScaler()
cluster_std = cluster_profile.copy()
cluster_std[categories] = scaler.fit_transform(cluster_profile[categories])

# จำนวนแกน
N = len(categories)
angles = np.linspace(0, 2 * np.pi, N, endpoint=False).tolist()
angles += angles[:1]  # ปิดวง

# สร้าง radar chart
fig, ax = plt.subplots(figsize=(6,6), subplot_kw=dict(polar=True))

for i, row in cluster_std.iterrows():
    values = row[categories].tolist()
    values += values[:1]  # ปิดวง

    # Set line width based on whether any value is not equal to 0
    line_width = 2 if any(v != 0 for v in values[:-1]) else 1

    ax.plot(angles, values, label=f"Cluster {row['Cluster_sd']}", linewidth=line_width)
    ax.fill(angles, values, alpha=0.25)

# ปรับ labels
ax.set_xticks(angles[:-1])
# Add padding to the labels to move them away from the center
ax.set_xticklabels(categories, ha='center', va='center', fontsize=10, color='dimgray') # Adjust ha, va, fontsize, color as needed
ax.tick_params(axis='x', pad=20) # Add padding to the x-axis ticks


plt.title("Standardized Cluster Profile Radar Chart")
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1.1))
plt.show()
````

### เหตุผลที่เลือกใช้ Random Forest Regressor

เราตัดสินใจเลือกใช้ Random Forest Regressor ในการทำนายยอดขายรายสัปดาห์จากหลายเหตุผล:

1.  **ความแม่นยำ:** Random Forest มักให้ผลการทำนายที่ค่อนข้างแม่นยำสำหรับงาน Regression โดยเฉพาะกับข้อมูลที่มีความสัมพันธ์ที่ซับซ้อน
2.  **จัดการกับความสัมพันธ์ที่ไม่เป็นเชิงเส้น:** ตัวแบบนี้สามารถจับรูปแบบความสัมพันธ์ระหว่างคุณลักษณะต่างๆ กับยอดขายที่ไม่ใช่เส้นตรงได้ดี ซึ่งเป็นเรื่องปกติในข้อมูลยอดขายจริง
3.  **ความทนทานต่อ Outliers:** Random Forest ไม่ค่อยอ่อนไหวต่อค่าผิดปกติ (Outliers) ในข้อมูลมากนัก เมื่อเทียบกับโมเดลประเภทอื่นๆ
4.  **จัดการข้อมูลได้หลากหลาย:** สามารถทำงานได้ดีกับคุณลักษณะทั้งที่เป็นตัวเลขและหมวดหมู่ (เมื่อแปลงแล้ว) โดยไม่ต้องปรับสเกลข้อมูล
5.  **บอกความสำคัญของ Feature:** ตัวแบบสามารถบอกได้ว่าคุณลักษณะใดมีความสำคัญที่สุดในการทำนาย ซึ่งช่วยให้เราเข้าใจปัจจัยขับเคลื่อนยอดขายได้ดียิ่งขึ้น
6.  **ลดปัญหา Overfitting:** การใช้เทคนิค Ensemble (รวมหลายต้นไม้) ช่วยลดความเสี่ยงที่โมเดลจะเรียนรู้ข้อมูลฝึกสอนมากเกินไปจนทำนายข้อมูลใหม่ได้ไม่ดี

````python
import pandas as pd # ตรวจสอบว่า import pandas แล้ว

# --- แก้ไขส่วนของ 'Type' ---
# Handle the 'Type' column
if 'Type' in df.columns:
    # เพิ่ม dtype=int เพื่อให้ผลลัพธ์เป็น 0/1 แทน True/False
    df = pd.get_dummies(df, columns=['Type'], prefix='Type', dtype=int)
    print("One-Hot Encoded 'Type' column (as 0/1).")
else:
    print("Warning: 'Type' column not found for encoding.")


# Now, handle the 'IsHoliday' column.
# (ส่วนนี้โค้ดเดิมของคุณจัดการดีแล้ว แต่เราจะใช้ dtype=int เพื่อให้สอดคล้องกัน)
if 'IsHoliday_y' in df.columns:
    df = df.drop('IsHoliday_y', axis=1)
    print("Dropped 'IsHoliday_y' column.")

# --- แก้ไขส่วนของ 'IsHoliday_x' ---
if 'IsHoliday_x' in df.columns:
    # เพิ่ม dtype=int และเราไม่จำเป็นต้องใช้ .astype(int) อีกต่อไป
    df = pd.get_dummies(df, columns=['IsHoliday_x'], prefix='IsHoliday', dtype=int)
    print("One-Hot Encoded 'IsHoliday_x' column (as 0/1).")

    # บรรทัด .astype(int) ด้านล่างนี้ไม่จำเป็นแล้ว เพราะ dtype=int จัดการให้แล้ว
    # if 'IsHoliday_False' in df.columns:
    #     df['IsHoliday_False'] = df['IsHoliday_False'].astype(int)
    # if 'IsHoliday_True' in df.columns:
    #     df['IsHoliday_True'] = df['IsHoliday_True'].astype(int)

# --- แก้ไขส่วนของ 'IsHoliday' (Fallback) ---
elif 'IsHoliday' in df.columns:
    # เพิ่ม dtype=int ที่นี่ด้วย
    df = pd.get_dummies(df, columns=['IsHoliday'], prefix='IsHoliday', dtype=int)
    print("One-Hot Encoded 'IsHoliday' column (as 0/1).")

    # บรรทัด .astype(int) ด้านล่างนี้ไม่จำเป็นแล้ว
    # if 'IsHoliday_False' in df.columns:
    #     df['IsHoliday_False'] = df['IsHoliday_False'].astype(int)
    # if 'IsHoliday_True' in df.columns:
    #     df['IsHoliday_True'] = df['IsHoliday_True'].astype(int)

else:
    print("Warning: Neither 'IsHoliday_x' nor 'IsHoliday' column found for encoding.")


print("\nข้อมูลหลัง One-Hot Encoding (5 แถวแรก):")
display(df.head())
````

ก่อนทำ One-Hot Encoding คอลัมน์ IsHoliday_x มีค่าอยู่ 2 แบบคือ True (เป็นวันหยุด) หรือ False (ไม่เป็นวันหยุด) เราต้องการแปลงข้อมูล 2 แบบนี้ให้เป็นตัวเลขครับ
One-Hot Encoding ทำแบบนี้:
สร้างคอลัมน์ใหม่ขึ้นมา 2 คอลัมน์ คือ IsHoliday_False และ IsHoliday_True
สำหรับแต่ละแถวข้อมูล:
* ถ้าค่าเดิมใน IsHoliday_x เป็น False (ไม่เป็นวันหยุด):

คอลัมน์ IsHoliday_False จะมีค่าเป็น 1

คอลัมน์ IsHoliday_True จะมีค่าเป็น 0
* ถ้าค่าเดิมใน IsHoliday_x เป็น True (เป็นวันหยุด):

คอลัมน์ IsHoliday_False จะมีค่าเป็น 0

คอลัมน์ IsHoliday_True จะมีค่าเป็น 1

ดังนั้น:

* ค่า 1 ในคอลัมน์ใดคอลัมน์หนึ่ง (เช่น IsHoliday_True = 1) หมายถึงว่าแถวข้อมูลนั้น "ตรงกับ" ประเภทที่คอลัมน์นั้นสื่อถึง
* ค่า 0 ในคอลัมน์ใดคอลัมน์หนึ่ง (เช่น IsHoliday_True = 0) หมายถึงว่าแถวข้อมูลนั้น "ไม่ตรงกับ" ประเภทที่คอลัมน์นั้น

คอลัมน์ที่สร้างขึ้นและความหมาย:

เหตุผล: ยอดขายมักมีรูปแบบตามฤดูกาล เช่น อาจจะสูงขึ้นในช่วงเทศกาลต่างๆ ที่อยู่ในสัปดาห์เฉพาะ การรู้หมายเลขสัปดาห์ช่วยให้โมเดลเรียนรู้รูปแบบเหล่านี้ได้
2. month: คอลัมน์นี้เก็บ หมายเลขเดือน ของปี 1 คือ มกราคม, 2 คือ กุมภาพันธ์, ... 12 คือ ธันวาคม
เหตุผล: คล้ายกับสัปดาห์ ยอดขายมีแนวโน้มตามเดือน เช่น อาจจะสูงขึ้นในช่วงปลายปี (เดือน 11, 12) เนื่องจากเทศกาลต่างๆ การรู้เดือนช่วยให้โมเดลจับแนวโน้มตามเดือนได้
3. year: คอลัมน์นี้เก็บ ปี ครับ เช่น 2010, 2011, 2012
เหตุผล: ช่วยให้โมเดลเรียนรู้แนวโน้มยอดขายในแต่ละปี ซึ่งอาจแตกต่างกันไปเนื่องจากปัจจัยทางเศรษฐกิจที่เปลี่ยนแปลงไปในแต่ละปี
4. dayofweek: คอลัมน์นี้เก็บ หมายเลขวันในสัปดาห์ โดยทั่วไป วันจันทร์คือ 0, อังคารคือ 1, ... อาทิตย์คือ 6
เหตุผล: ยอดขายในแต่ละวันของสัปดาห์อาจแตกต่างกัน เช่น ยอดขายช่วงสุดสัปดาห์อาจจะสูงกว่าวันธรรมดา การรู้หมายเลขวันช่วยให้โมเดลพิจารณาความแตกต่างนี้ได้
5. quarter: คอลัมน์นี้เก็บ หมายเลขไตรมาส ของปีครับ ไตรมาส 1 คือ ม.ค.-มี.ค., ไตรมาส 2 คือ เม.ย.-มิ.ย., ไตรมาส 3 คือ ก.ค.-ก.ย., ไตรมาส 4 คือ ต.ค.-ธ.ค.
เหตุผล: บางธุรกิจมียอดขายที่แตกต่างกันอย่างชัดเจนในแต่ละไตรมาส (เช่น ไตรมาส 4 ที่มีเทศกาลสำคัญเยอะ) การรู้ไตรมาสช่วยให้โมเดลเรียนรู้รูปแบบตามไตรมาสได้

````python
# Select features (X) and target (y)
# เราจะเลือกคอลัมน์ทั้งหมด ยกเว้น 'Weekly_Sales' (ที่เราต้องการทำนาย)
# และคอลัมน์ 'Date' รวมถึงคอลัมน์ที่ซ้ำซ้อนหรือถูกแปลงไปแล้ว

# Define columns to drop initially
cols_to_drop = ['Weekly_Sales','Date']

# Check if 'IsHoliday_y' exists and add to cols_to_drop if it does
if 'IsHoliday_y' in df.columns:
    cols_to_drop.append('IsHoliday_y')
    print("Dropping 'IsHoliday_y' column as it exists.")
else:
    print("'IsHoliday_y' column not found, skipping drop.")

# Drop the selected columns from df to create X
X = df.drop(cols_to_drop, axis=1)
y = df['Weekly_Sales']

print("ตัวแปรอิสระ (X) 5 แถวแรก:")
display(X.head())

print("\nตัวแปรตาม (y) 5 แถวแรก:")
display(y.head())
````

### ขั้นตอนที่ 5: การสร้างและฝึกตัวแบบ (Build and Train the Model)

สร้าง Object ของ Random Forest Regressor และทำการฝึกด้วยข้อมูล `X_train` และ `y_train`

````python
from sklearn.ensemble import RandomForestRegressor

# Instantiate the model
# n_estimators คือจำนวนต้นไม้ในป่า (ลองเริ่มที่ 100)
# random_state ช่วยให้ผลการฝึกเหมือนเดิมทุกครั้งที่รัน
# n_jobs=-1 ใช้ CPU ทุก Core เพื่อให้ฝึกเร็วขึ้น
model = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)

# Train the model
model.fit(X_train, y_train)

print("Random Forest Regressor Model training complete.")
````

1. n_estimators=100
อธิบาย: พารามิเตอร์นี้กำหนด จำนวนต้นไม้ตัดสินใจ ที่จะสร้างขึ้นในป่า Random Forest ครับ Random Forest ทำงานโดยการรวมผลการทำนายจากต้นไม้หลายๆ ต้นเข้าด้วยกัน ยิ่งมีต้นไม้เยอะ โมเดลก็มักจะยิ่งแม่นยำและมีความทนทานต่อ Overfitting มากขึ้น

ทำไมใช้ 100: ค่า 100 เป็นค่าเริ่มต้นที่นิยมใช้และมักจะให้ผลลัพธ์ที่ดีในหลายๆ กรณีครับ การเพิ่มจำนวนต้นไม้ให้มากขึ้นกว่านี้ (เช่น 200, 500, 1000) อาจช่วยเพิ่มความแม่นยำได้อีกเล็กน้อย แต่ก็ต้องใช้เวลาในการฝึกและประมวลผลนานขึ้น เราสามารถลองปรับค่านี้เพื่อหาจำนวนต้นไม้ที่เหมาะสมกับข้อมูลและทรัพยากรที่เรามีได้

2. random_state=42
อธิบาย: พารามิเตอร์นี้ใช้สำหรับกำหนด ค่าเริ่มต้นของตัวสร้างเลขสุ่ม (Random Number Generator) ครับ ใน Random Forest มีกระบวนการสุ่มอยู่หลายส่วน เช่น การสุ่มเลือกข้อมูลย่อยเพื่อสร้างแต่ละต้นไม้ และการสุ่มเลือกคุณลักษณะที่จะใช้ในการแบ่งข้อมูล
ทำไมใช้ 42: การกำหนด random_state เป็นค่าคงที่ (เช่น 42 หรือเลขอื่นๆ ที่คุณชอบ) จะทำให้ ผลลัพธ์ของการฝึกโมเดลเหมือนเดิมทุกครั้ง ที่คุณรันโค้ดครับ หากไม่กำหนดค่านี้ ทุกครั้งที่รันโค้ด โมเดลอาจจะได้ต้นไม้ที่แตกต่างกันเล็กน้อยเนื่องจากการสุ่ม ทำให้ผลการทำนายอาจเปลี่ยนแปลงไปได้
เหตุผล: การกำหนด random_state มีประโยชน์มากในการทดลองและพัฒนาโมเดลครับ เพราะช่วยให้เรามั่นใจได้ว่าการเปลี่ยนแปลงใดๆ ในโค้ด (เช่น การเตรียมข้อมูล หรือการปรับ Hyperparameters อื่นๆ) คือสาเหตุที่แท้จริงที่ทำให้ผลลัพธ์เปลี่ยนไป ไม่ใช่เพราะการสุ่มที่แตกต่างกันในแต่ละครั้งครับ เลข 42 เป็นเลขที่นิยมใช้ในวงการ Data Science

3. n_jobs=-1
อธิบาย: พารามิเตอร์นี้กำหนดว่าเราต้องการให้โมเดลใช้ จำนวน Core ของ CPU ในการทำงานเท่าไหร่ครับ
ทำไมใช้ -1: ค่า -1 เป็นค่าพิเศษที่หมายถึง ให้ใช้ Core ของ CPU ทั้งหมดที่มีอยู่ ครับ การฝึกโมเดล Random Forest สามารถทำแบบขนานได้ (ฝึกแต่ละต้นไม้แยกกัน) การใช้ Core ทั้งหมดจะช่วยให้กระบวนการฝึกโมเดลเสร็จเร็วขึ้นอย่างมาก โดยเฉพาะเมื่อมีจำนวนต้นไม้เยอะๆ หรือข้อมูลมีขนาดใหญ่ค

**สรุปคือ** พารามิเตอร์เหล่านี้ช่วยให้เราควบคุมลักษณะของโมเดล Random Forest ครับ n_estimators กำหนดขนาดของป่า (จำนวนต้นไม้), random_state ช่วยให้ผลลัพธ์คงที่สำหรับการทดลอง และ n_jobs ช่วยให้การฝึกโมเดลเร็วขึ้นครับ ค่าที่ใช้ (100, 42, -1) เป็นค่าเริ่มต้นที่นิยมและเหมาะสมสำหรับการเริ่มต้น

### ขั้นตอนที่ 6: การประเมินประสิทธิภาพของตัวแบบ (Model Evaluation)

ใช้ Metric มาตรฐานสำหรับ Regression (MAE, MSE, RMSE) ในการวัดว่าตัวแบบทำนายได้แม่นยำแค่ไหนเมื่อเทียบกับค่าจริงในข้อมูลชุดทดสอบครับ

*   **Histogram of Residuals:** หากข้อผิดพลาดเป็นการสุ่ม การกระจายตัวของ Residuals ควรมีลักษณะใกล้เคียง Normal Distribution (รูประฆังคว่ำ) ที่มีค่าเฉลี่ยใกล้เคียงศูนย์ครับ การที่กราฟเบ้ไปทางใดทางหนึ่ง หรือมีหลายยอด อาจบ่งชี้ถึงปัญหาในการทำนายครับ

### ขั้นตอนที่ 9: การทำนายข้อมูลทดสอบ (Prediction on Test Data)

เตรียมข้อมูลจาก `test.csv` ให้มีรูปแบบเหมือนกับข้อมูลฝึกสอน (`X_train`) และใช้ตัวแบบ Random Forest Regressor ที่ฝึกแล้วในการทำนายยอดขาย

````python
# Load the actual test data
test_df = pd.read_csv(test_path)

# Merge with features and stores data, similar to how the training data was merged
test_df = test_df.merge(df_ft, on=['Store', 'Date'], how='inner').merge(df_st, on=['Store'], how='inner')

# Handle missing values in MarkDown columns (fill with 0)
markdown_cols = ['MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4', 'MarkDown5']
test_df[markdown_cols] = test_df[markdown_cols].fillna(0)

# Create date-related features
test_df["Date"] = pd.to_datetime(test_df["Date"])
test_df['week'] = test_df['Date'].dt.isocalendar().week.astype(int)
test_df['month'] = test_df['Date'].dt.month
test_df['year'] = test_df['Date'].dt.year
test_df['dayofweek'] = test_df['Date'].dt.dayofweek
test_df['quarter'] = test_df['Date'].dt.quarter

# Perform One-Hot Encoding for 'Type' and 'IsHoliday' (Note: test.csv uses 'IsHoliday', not 'IsHoliday_x')
test_df = pd.get_dummies(test_df, columns=['Type'], prefix='Type')
if 'IsHoliday' in test_df.columns: # Use 'IsHoliday' column from test.csv
    test_df = pd.get_dummies(test_df, columns=['IsHoliday'], prefix='IsHoliday_x') # Rename to match training data column prefix

# Drop the original 'Date' column and potentially redundant 'IsHoliday' column if it exists and wasn't encoded
test_df = test_df.drop(['Date'], axis=1)
if 'IsHoliday' in test_df.columns: # Drop original 'IsHoliday' if still exists after encoding
    test_df = test_df.drop(['IsHoliday'], axis=1)


# Ensure test_df has the exact same columns as X_train, fill missing columns with 0 if any
missing_cols_in_test = set(X_train.columns) - set(test_df.columns)
for c in missing_cols_in_test:
    test_df[c] = 0

# Ensure the order of columns in test_df is the same as X_train
test_df = test_df[X_train.columns]

print("Processed test data head (ready for prediction):")
display(test_df.head())
````

### ขั้นตอนที่ 13: การประเมินประสิทธิภาพของตัวแบบรอบที่สอง (Model Evaluation - Selected Features)

ใช้ Metric มาตรฐาน (MAE, MSE, RMSE) ในการวัดว่าตัวแบบ Random Forest รอบที่สองที่ใช้คุณลักษณะที่เลือกแล้ว ทำนายได้แม่นยำแค่ไหนเมื่อเทียบกับค่าจริงในข้อมูลชุดทดสอบ

````python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Ensure model_selected_features, X_test_selected, and y_test_selected are defined
if 'model_selected_features' in locals() and 'X_test_selected' in locals() and 'y_test_selected' in locals():
    # Make predictions on the test set with selected features
    predictions_selected = model_selected_features.predict(X_test_selected)

    # Evaluate the second model
    mae_selected = mean_absolute_error(y_test_selected, predictions_selected)
    mse_selected = mean_squared_error(y_test_selected, predictions_selected)
    rmse_selected = np.sqrt(mse_selected)

    print(f"Mean Absolute Error (MAE) (Selected Features): {mae_selected}")
    print(f"Mean Squared Error (MSE) (Selected Features): {mse_selected}")
    print(f"Root Mean Squared Error (RMSE) (Selected Features): {rmse_selected}")

    print("\n====== สรุปผลการประเมินตัวแบบรอบสอง: ประเมินสำเร็จ ======")

    # Optional: Compare with the first model's performance
    if 'mae' in locals() and 'mse' in locals() and 'rmse' in locals():
        print("\n====== การเปรียบเทียบกับตัวแบบแรก (ใช้ Feature ทั้งหมด) ======")
        print(f"ตัวแบบแรก: MAE={mae}, MSE={mse}, RMSE={rmse}")
        print(f"ตัวแบบรอบสอง (Selected Features): MAE={mae_selected}, MSE={mse_selected}, RMSE={rmse_selected}")
        print("==========================================================")
        if mae_selected < mae:
            print(">>> MAE ลดลง: การเลือก Feature ช่วยให้ความคลาดเคลื่อนเฉลี่ยน้อยลง")
        elif mae_selected > mae:
             print(">>> MAE เพิ่มขึ้น: การเลือก Feature อาจทำให้ความคลาดเคลื่อนเฉลี่ยมากขึ้น")
        else:
             print(">>> MAE เท่าเดิม")

        if rmse_selected < rmse:
            print(">>> RMSE ลดลง: การเลือก Feature ช่วยลดความคลาดเคลื่อนรวม")
        elif rmse_selected > rmse:
             print(">>> RMSE เพิ่มขึ้น: การเลือก Feature อาจทำให้ความคลาดเคลื่อนรวมมากขึ้น")
        else:
             print(">>> RMSE เท่าเดิม")
    else:
        print("\nไม่พบ Metric ของตัวแบบแรก ไม่สามารถเปรียบเทียบได้")


else:
    print("ตัวแปรโมเดลรอบสอง หรือข้อมูลทดสอบที่เลือก Feature ยังไม่ได้ถูกสร้างขึ้น โปรดตรวจสอบขั้นตอนที่ 11 และ 12")
````

## ประโยชน์ของการทำนายยอดขายรายสัปดาห์

การสร้างแบบจำลองเพื่อทำนายยอดขายรายสัปดาห์มีความสำคัญและมีประโยชน์อย่างยิ่งต่อธุรกิจค้าปลีก เช่น Walmart ในหลายๆ ด้าน ดังนี้ครับ:

1.  **การบริหารจัดการสินค้าคงคลัง (Inventory Management):** การพยากรณ์ยอดขายที่แม่นยำช่วยให้ร้านค้าสามารถวางแผนการสั่งซื้อและจัดเก็บสินค้าได้อย่างเหมาะสม ลดปัญหาการมีสินค้ามากเกินไป (Overstocking) ซึ่งทำให้เกิดต้นทุนการจัดเก็บสูง และลดปัญหาการขาดแคลนสินค้า (Stockouts) ซึ่งทำให้เสียโอกาสในการขาย
2.  **การจัดสรรทรัพยากร:** ช่วยในการวางแผนจัดสรรทรัพยากรต่างๆ เช่น จำนวนพนักงานที่ต้องการในแต่ละสัปดาห์ โดยเฉพาะในช่วงที่มีการคาดการณ์ยอดขายสูง เพื่อให้การบริการลูกค้าเป็นไปอย่างราบรื่นและมีประสิทธิภาพ
3.  **การวางแผนการตลาดและโปรโมชั่น:** การเข้าใจว่าปัจจัยใด (เช่น วันหยุด, โปรโมชั่น) มีผลต่อยอดขายมากที่สุด ช่วยให้ทีมการตลาดสามารถวางแผนแคมเปญและโปรโมชั่นได้อย่างมีประสิทธิภาพ กำหนดช่วงเวลาและประเภทของโปรโมชั่นที่เหมาะสมเพื่อกระตุ้นยอดขายสูงสุด
4.  **การคาดการณ์และประเมินผลประกอบการ:** ช่วยให้ผู้บริหารสามารถคาดการณ์แนวโน้มยอดขายในอนาคตได้ ทำให้วางแผนงบประมาณ, เป้าหมายรายได้, และประเมินผลประกอบการของร้านค้าและแผนกต่างๆ ได้อย่างมีข้อมูลสนับสนุน
5.  **การตัดสินใจเชิงกลยุทธ์:** ข้อมูลจากการทำนายยอดขายและ Feature Importance ช่วยให้เข้าใจปัจจัยขับเคลื่อนยอดขาย ทำให้สามารถตัดสินใจเชิงกลยุทธ์ได้ดีขึ้น เช่น การขยายร้านค้า, การปรับเปลี่ยนกลยุทธ์ด้านราคา, หรือการปรับปรุงการจัดวางสินค้าในร้าน
