````python
# =========================
# Auto ARIMA (SARIMA search) - concise
# =========================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pmdarima import auto_arima
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error

# 0) โหลด/เตรียม series (ปรับ path ถ้าจำเป็น)
df = pd.read_csv("/content/drive/MyDrive/Walmart/time_series.csv")
df['Date'] = pd.to_datetime(df['Date'])
series = df.groupby('Date')['Weekly_Sales'].sum().sort_index()
# resample ให้เป็นสัปดาห์ที่ตรงกับ data (เช่น 'W-FRI' ถ้าบันทึกทุกศุกร์)
series = series.resample('W-FRI').sum().interpolate(method='time')

# 1) แบ่ง train/test (time-based)
split = int(len(series) * 0.7)
train = series.iloc[:split]
test  = series.iloc[split:]

# 2) เรียก auto_arima (search เฉพาะ seasonal models, m=52)
# ปรับ start/max params ตามต้องการ (ไม่ให้ search กว้างเกินไปเพื่อความเร็ว)
auto_model = auto_arima(
    train,
    start_p=0, start_q=0,
    max_p=3, max_q=3,
    d=None,            # ให้ auto_arima เลือก d อัตโนมัติ
    seasonal=True,
    start_P=0, start_Q=0,
    max_P=2, max_Q=2,
    D=None,            # ให้ auto_arima เลือก D อัตโนมัติ
    m=52,              # seasonal period = 52 สัปดาห์
    trace=True,
    error_action='ignore',
    suppress_warnings=True,
    stepwise=True,
    n_jobs=1
)

print("\n=== Auto ARIMA summary ===")
print(auto_model.summary())

# 3) พยากรณ์ช่วง test (ใช้ model ที่ได้)
n_test = len(test)
pred_test = auto_model.predict(n_periods=n_test)
pred_index = test.index

# 4) Metrics
rmse = np.sqrt(mean_squared_error(test.values, pred_test))
mape = mean_absolute_percentage_error(test.values, pred_test) * 100
print(f"\nRMSE: {rmse:,.2f}")
print(f"MAPE: {mape:.2f}%")

# 5) Plot Actual vs Predicted
plt.figure(figsize=(12,5))
plt.plot(train.index, train.values, label='Train', color='steelblue')
plt.plot(test.index, test.values, label='Actual (Test)', color='orange')
plt.plot(pred_index, pred_test, label='Predicted (auto_arima)', color='green', linestyle='--')
plt.title('Auto ARIMA (SARIMA search) — Actual vs Predicted')
plt.xlabel('Date'); plt.ylabel('Weekly Sales')
plt.legend(); plt.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
````

````python
import matplotlib.pyplot as plt
import pandas as pd

# Calculate average sales for each specific holiday and non-holiday
avg_sales_by_holiday_type = df.groupby('Super_Bowl')['Weekly_Sales'].mean().rename('Super_Bowl').to_frame()
avg_sales_by_holiday_type['Labor_Day'] = df.groupby('Labor_Day')['Weekly_Sales'].mean()
avg_sales_by_holiday_type['Black_Friday'] = df.groupby('Black_Friday')['Weekly_Sales'].mean()
avg_sales_by_holiday_type["New_Year's_Eve"] = df.groupby("New_Year's_Eve")['Weekly_Sales'].mean()

# Transpose the DataFrame to have holidays as rows and True/False as columns
avg_sales_by_holiday_type = avg_sales_by_holiday_type.T

# Rename columns for clarity
avg_sales_by_holiday_type.rename(columns={False: 'Non_Holiday', True: 'Holiday'}, inplace=True)

# Display the calculated averages
print("Average Weekly Sales by Specific Holiday Type:")
display(avg_sales_by_holiday_type)

# Plotting
fig, ax = plt.subplots(figsize=(25, 15))

# Get the index (holiday names) and the values (sales)
holiday_names = avg_sales_by_holiday_type.index
non_holiday_sales = avg_sales_by_holiday_type['Non_Holiday']
holiday_sales = avg_sales_by_holiday_type['Holiday']

bar_width = 0.35
x_pos = range(len(holiday_names))

rects1 = ax.bar(x_pos, holiday_sales, bar_width, label='True', color='blue')
rects2 = ax.bar([p + bar_width for p in x_pos], non_holiday_sales, bar_width, label='False', color='red')


# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Average Weekly Sales')
ax.set_title('Average Weekly Sales by Specific Holiday Type')
ax.set_xticks([p + bar_width / 2 for p in x_pos])
ax.set_xticklabels(holiday_names)
ax.legend()

def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:,.0f}',  # Format sales as integers with comma
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
````

ความสำคัญของปัญหา

การพยากรณ์ยอดขายรายสัปดาห์ที่แม่นยำเป็นหัวใจสำคัญของการดำเนินงานร้านค้าปลีก:

- การบริหารสินค้าคงคลัง (Inventory Management): ช่วยลดต้นทุนการเก็บรักษาสินค้า (Holding Costs) และป้องกันการขาดแคลนสินค้า (Stockouts)

- การจัดสรรบุคลากร: วางแผนกำลังคนให้เพียงพอในช่วงยอดขายสูง

- การวางแผนกลยุทธ์การตลาด: เข้าใจว่าโปรโมชั่น (MarkDowns) ใดมีประสิทธิภาพสูงสุด

ข้อคาดการณ์ (Conjectures) 3 ข้อ

1. ข้อคาดการณ์ที่ 1 (วันหยุด): ยอดขายรายสัปดาห์ในช่วงสัปดาห์ที่เป็นวันหยุด (โดยเฉพาะวันหยุดใหญ่) จะมีค่าเฉลี่ย สูงกว่า สัปดาห์ปกติอย่างมีนัยสำคัญ

2. ข้อคาดการณ์ที่ 2 (CPI/Unemployment): ตัวชี้วัดทางเศรษฐกิจ CPI (เงินเฟ้อ) และ Unemployment (ว่างงาน) จะมีความสัมพันธ์ เชิงลบ กับยอดขาย (เมื่อเศรษฐกิจไม่ดี ยอดขายจะลดลง)

3. ข้อคาดการณ์ที่ 3 (MarkDowns): โปรโมชั่นลดราคา MarkDowns จะมีความสัมพันธ์ เชิงบวก กับยอดขาย แต่ MarkDown บางประเภทจะมีประสิทธิภาพสูงกว่าประเภทอื่น

````python
import pandas as pd
from statsmodels.tsa.stattools import adfuller

# 1. โหลดและเตรียมข้อมูล (ต้องทำก่อน ADF Test)
df['Date'] = pd.to_datetime(df['Date'])

# สร้าง Time Series โดยการรวมยอดขายทั้งหมด
# โค้ดนี้จะสร้าง Pandas Series ที่มี Date เป็น Index และ Weekly_Sales เป็นค่า
weekly_sales_series = df.groupby('Date')['Weekly_Sales'].sum()

# 2. ใช้ ADF Test
# ต้องใส่ Weekly_Sales Series เข้าไป (ไม่ใช่ Date)
adf_test = adfuller(weekly_sales_series)

# 3. Output the results
print('ADF Statistic: %f' % adf_test[0])
print('p-value: %f' % adf_test[1])
````

````python
#Auto Regressive (p)
from statsmodels.graphics.tsaplots import  plot_pacf
from statsmodels.tsa.stattools import pacf

plt.rcParams.update({'figure.figsize':(10,4)})
plot_pacf(weekly_sales_series.dropna(), method = 'ols')
df_pacf = pacf(weekly_sales_series.dropna(), method = 'ols')
````

````python
import numpy as np

for i in range(0, len(df_pacf)):
    # Corrected line to use the length of the time series data
    if df_pacf[i] < 1.96 / np.sqrt(len(weekly_sales_series.dropna())):
        print('p= ', i - 1)
        break
````

````python
#Moving Average (q)
from statsmodels.graphics.tsaplots import  plot_acf
from statsmodels.tsa.stattools import acf

plt.rcParams.update({'figure.figsize':(10,4)})
plot_acf(weekly_sales_series, fft = True)
df_acf = acf(weekly_sales_series, fft = True)
````

````python

for i in range(0, len(df_acf)):
    if df_acf[i] < 1.96 / np.sqrt(len(weekly_sales_series)):
        print('q=', i - 1)
        break
````

````python
#Test order
import itertools

p = range(1, 2)
d = range(0, 1)
q = range(0, 3)
pdq = list(itertools.product(p, d, q))
print(pdq)

from statsmodels.tsa.arima_model import ARIMA

aic = []

for param in pdq:
    try:
        model = ARIMA(weekly_sales_series.dropna(), order = param)
        results = model.fit()
        print('Order = {}'.format(param))
        print('AIC = {}'.format(results.aic))
        a = 'Order: '+str(param) +' AIC: ' + str(results.aic)
        aic.append(a)
    except:
        continue
````

## เตรียมข้อมูล (data preparation)

### Subtask:
เลือกคุณลักษณะที่เกี่ยวข้อง เช่น Weekly_Sales, Size, และอาจรวมถึงปัจจัยภายนอกอื่นๆ ที่อาจส่งผลต่อประเภทของร้านค้า. เตรียมข้อมูลสำหรับนำไปใช้ในการสร้างโมเดล Classification.


**Reasoning**:
Select the relevant features and the target variable for the classification task.



````python
features = ['Weekly_Sales', 'Size', 'IsHoliday', 'Temperature', 'Fuel_Price', 'CPI', 'Unemployment', 'MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4', 'MarkDown5', 'week', 'month', 'year']
X = df[features]
y = df['Type']
````

## แบ่งชุดข้อมูล (data splitting)

### Subtask:
แบ่งข้อมูลออกเป็นชุดฝึก (Training set) และชุดทดสอบ (Testing set) เพื่อใช้ในการสร้างและประเมินตัวแบบ.


**Reasoning**:
Split the feature and target data into training and testing sets for model training and evaluation.



````python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
````

## สร้างและฝึกตัวแบบ (model building and training)

### Subtask:
เลือกใช้โมเดล Classification ที่เหมาะสมกับข้อมูล Multiclass เช่น Logistic Regression, Decision Tree, Random Forest, หรือ Support Vector Machine (SVM) แล้วทำการฝึกโมเดลโดยใช้ชุดข้อมูลฝึก (Training set) ที่ได้จากการแบ่งชุดข้อมูล


**Reasoning**:
Import the RandomForestClassifier model and train it using the training data.



````python
from sklearn.ensemble import RandomForestClassifier

model_rf = RandomForestClassifier(n_estimators=100, random_state=42)
model_rf.fit(X_train, y_train)
````

## ประเมินผลตัวแบบ (model evaluation)

### Subtask:
ประเมินประสิทธิภาพของโมเดล Random Forest Classifier ที่สร้างขึ้นสำหรับการจำแนกประเภทของร้านค้าโดยใช้ชุดข้อมูลทดสอบ (Testing set) และ Metric ที่เหมาะสม เช่น Accuracy, Precision, Recall, F1-Score, และ Confusion Matrix.


**Reasoning**:
Evaluate the trained Random Forest Classifier model using the testing set and relevant classification metrics, and display the results.



````python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import pandas as pd

# Predict the store types on the test set
y_pred = model_rf.predict(X_test)

# Calculate Accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Display Classification Report
report = classification_report(y_test, y_pred, output_dict=True)
df_report = pd.DataFrame(report).transpose()
print("\nClassification Report:")
display(df_report)

# Display Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
df_cm = pd.DataFrame(cm, index=['Actual A', 'Actual B', 'Actual C'], columns=['Predicted A', 'Predicted B', 'Predicted C'])
print("\nConfusion Matrix:")
display(df_cm)
````

## วิเคราะห์คุณลักษณะที่สำคัญ (feature importance - optional)

### Subtask:
หากใช้โมเดลที่รองรับ (เช่น Tree-based models) สามารถวิเคราะห์ได้ว่าคุณลักษณะใดมีความสำคัญที่สุดในการจำแนกประเภทของร้านค้า


**Reasoning**:
Extract feature importances from the trained Random Forest model, create a DataFrame to store feature names and importances, sort the DataFrame, and display it. Then, create a horizontal bar plot to visualize feature importances.



````python
import matplotlib.pyplot as plt

# 1. Access feature importances
importances = model_rf.feature_importances_
feature_names = X.columns

# 2. Create a DataFrame for feature importances
feature_importance_df = pd.DataFrame({'feature': feature_names, 'importance': importances})

# 3. Sort the DataFrame by importance
feature_importance_df = feature_importance_df.sort_values('importance', ascending=False)

# 4. Display the sorted DataFrame
print("Feature Importances:")
display(feature_importance_df)

# 5. Create a horizontal bar plot
plt.figure(figsize=(10, 7))
plt.barh(feature_importance_df['feature'], feature_importance_df['importance'], color='skyblue')
plt.title("Feature Importance for Store Type Classification (Random Forest)")
plt.xlabel("Importance Score")
plt.ylabel("Feature")
plt.gca().invert_yaxis() # Invert y-axis to have the most important feature at the top
plt.show()
````

#### ปัญหา/คำถามที่สนใจมีความสำคัญอย่างไร?

การจำแนกสัปดาห์ที่มียอดขายสูง (สัปดาห์ขายดี) มีความสำคัญอย่างยิ่งต่อการดำเนินงานของธุรกิจค้าปลีก เช่น Walmart ดังนี้:

1.  **การบริหารจัดการสินค้าคงคลัง (Inventory Management):** การรู้ล่วงหน้าว่าสัปดาห์ใดมีแนวโน้มที่ยอดขายจะสูง ช่วยให้ร้านค้าสามารถวางแผนการสั่งซื้อและจัดเก็บสินค้าได้อย่างมีประสิทธิภาพ ลดปัญหาการขาดแคลนสินค้า (Stockouts) ในช่วงที่ลูกค้าต้องการมาก และลดต้นทุนการเก็บรักษาสินค้า (Holding Costs) ในช่วงที่ยอดขายปกติ
2.  **การจัดสรรทรัพยากร (Resource Allocation):** ช่วยในการวางแผนกำลังคน (Staffing) ให้เพียงพอต่อความต้องการของลูกค้าในช่วงสัปดาห์ขายดี ทำให้บริการลูกค้าเป็นไปอย่างราบรื่น และจัดสรรทรัพยากรอื่นๆ เช่น งบประมาณการตลาด หรือการจัดเรียงสินค้า (Merchandising) ได้เหมาะสม
3.  **การวางแผนโปรโมชั่นและการตลาด (Promotion and Marketing Planning):** เข้าใจว่าปัจจัยใดที่ส่งผลให้เกิดสัปดาห์ขายดี ช่วยในการวางแผนกิจกรรมส่งเสริมการขายให้ตรงกับช่วงเวลาที่เหมาะสม เพื่อกระตุ้นยอดขายให้สูงยิ่งขึ้นไปอีก หรือรักษาโมเมนตัมของยอดขาย
4.  **การปรับปรุงกลยุทธ์การดำเนินงาน:** การวิเคราะห์ว่าปัจจัยใดมีอิทธิพลต่อสัปดาห์ขายดี ช่วยให้ธุรกิจเข้าใจพฤติกรรมผู้บริโภคและแนวโน้มตลาดได้ดียิ่งขึ้น นำไปสู่การปรับปรุงกลยุทธ์การดำเนินงานในระยะยาว

#### ข้อคาดการณ์ของท่านจะนำไปสู่การแก้ไขปัญหาหรือตอบคำถาม

จากคำถามและข้อมูลที่มี สามารถตั้งข้อคาดการณ์ (Conjectures) ได้ดังนี้ (ตัวอย่าง):

1.  **ข้อคาดการณ์ที่ 1 (วันหยุด):** สัปดาห์ที่มีวันหยุดสำคัญ (เช่น Black Friday, Super Bowl) จะมีแนวโน้มที่จะเป็นสัปดาห์ขายดีมากกว่าสัปดาห์ปกติ
2.  **ข้อคาดการณ์ที่ 2 (ปัจจัยเศรษฐกิจ):** เมื่อดัชนีราคาผู้บริโภค (CPI) ต่ำ หรืออัตราการว่างงาน (Unemployment) ต่ำลง (บ่งชี้เศรษฐกิจดีขึ้น) สัปดาห์นั้นๆ อาจมีแนวโน้มเป็นสัปดาห์ขายดีมากขึ้น
3.  **ข้อคาดการณ์ที่ 3 (อุณหภูมิ):** อุณหภูมิที่เหมาะสม (ไม่หนาวหรือร้อนจนเกินไป) อาจส่งผลให้ลูกค้าออกมาจับจ่ายใช้สอยมากขึ้น และทำให้สัปดาห์นั้นเป็นสัปดาห์ขายดีได้
4.  **(Optional) ข้อคาดการณ์ที่ 4 (MarkDowns):** การมีโปรโมชั่นลดราคา (MarkDowns) ในสัปดาห์นั้นๆ จะเพิ่มโอกาสที่สัปดาห์นั้นจะเป็นสัปดาห์ขายดี

ข้อคาดการณ์เหล่านี้จะถูกทดสอบและสนับสนุนหรือปฏิเสธด้วยการวิเคราะห์ข้อมูลและการสร้างโมเดล Classification ค่ะ หากโมเดลสามารถจำแนกสัปดาห์ขายดีได้แม่นยำ และ Feature Importance ชี้ว่าคุณลักษณะที่เกี่ยวข้องกับข้อคาดการณ์มีความสำคัญ ก็จะเป็นการสนับสนุนข้อคาดการณ์นั้นๆ ซึ่งจะนำไปสู่การตัดสินใจทางธุรกิจที่อิงจากข้อมูลค่ะ

````python
output_path = "/content/drive/MyDrive/Walmart/processed_walmart_train_data.csv"
df = pd.read_csv(output_path)
df.head()
````

#### สร้าง Target Variable (High_Sales)

````python
#กำหนดให้สัปดาห์ที่ยอดขายสูงกว่าค่าเฉลี่ยของร้านนั้นเป็น “High Sales”
# คำนวณยอดขายเฉลี่ยต่อร้าน
avg_sales_per_store = df.groupby('Store')['Weekly_Sales'].transform('mean')

# สร้าง target
df['High_Sales'] = (df['Weekly_Sales'] > avg_sales_per_store).astype(int)
````

####เลือกฟีเจอร์

````python
features = ['IsHoliday', 'Temperature', 'Fuel_Price', 'CPI', 'Unemployment', 'Type', 'Size']
X = df[features]
y = df['High_Sales']
````

````python
#แปลง Type (A/B/C) ให้เป็นตัวเลข
X = pd.get_dummies(X, columns=['Type'], drop_first=True)
````

````python
X
````

####แบ่งชุดข้อมูล

````python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
````

####สร้างและเทรนโมเดล

#####Logistic Regression

````python
from sklearn.linear_model import LogisticRegression

model_lr = LogisticRegression(max_iter=1000)
model_lr.fit(X_train, y_train)
````

#####Random Forest Classifier

````python
from sklearn.ensemble import RandomForestClassifier

model_rf = RandomForestClassifier(n_estimators=100, random_state=42)
model_rf.fit(X_train, y_train)
````

####ประเมินผล

````python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import pandas as pd

y_pred = model_rf.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))

# Display classification report as a DataFrame
report = classification_report(y_test, y_pred, output_dict=True)
df_report = pd.DataFrame(report).transpose()
print("\nClassification Report:")
display(df_report)

# Display confusion matrix as a DataFrame
cm = confusion_matrix(y_test, y_pred)
df_cm = pd.DataFrame(cm, index=['Actual 0', 'Actual 1'], columns=['Predicted 0', 'Predicted 1'])
print("\nConfusion Matrix:")
display(df_cm)
````

####วิเคราะห์ความสำคัญของฟีเจอร์

````python
import matplotlib.pyplot as plt

#ใช้ค่า Feature Importance เพื่อดูว่าตัวแปรใดมีผลต่อการเป็น “สัปดาห์ขายดี” มากที่สุด
importances = model_rf.feature_importances_
feature_names = X.columns

plt.figure(figsize=(8,5))
plt.barh(feature_names, importances)
plt.title("Feature Importance (Random Forest)")
plt.xlabel("Importance Score")
plt.ylabel("Feature")
plt.show()
````

6.  **การวิเคราะห์เชิงลึกอื่นๆ (Further Analysis):** การมีข้อมูลประเภทของร้านค้าที่แม่นยำเป็นพื้นฐานสำคัญสำหรับการวิเคราะห์เชิงลึกอื่นๆ เช่น การศึกษาผลกระทบของปัจจัยภายนอก (เศรษฐกิจ วันหยุด อุณหภูมิ) ต่อร้านค้าแต่ละประเภทโดยเฉพาะ
