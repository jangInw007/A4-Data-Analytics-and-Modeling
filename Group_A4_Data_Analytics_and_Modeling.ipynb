````python
#import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
````

#### Partition data
เลือกเฉพาะ column ที่สนใจ

````python
df1 = df[['Store','Size','Weekly_Sales','CPI','Unemployment','Fuel_Price']]
df1
````

````python
# Calculate the mean of all columns in df1, grouped by Store
df2 = df1.groupby('Store').median()

# Display the result
display(df2.reset_index().head())
````

````python
plt.scatter(x = df2['Size'], y = df2['Weekly_Sales'])
plt.xlabel('Size')
plt.ylabel('Weekly_Sales')
````

````python
df2['Sales_Per_Size'] = df2['Weekly_Sales'] / df2['Size']
````

````python
df_store_median = df2.drop(columns = ['Size','Weekly_Sales'])
df_store_median.reset_index().head()
````

````python
features = [
    'CPI',
    'Unemployment',
    'Fuel_Price',
    'Sales_Per_Size'
      # 'Weekly_Sales',
    # 'Size'
    ]

X = df_store_median[features]
````

#### Normalize

````python
#ปรับ scale ข้อมูลให้อยู่ในหน่วยเดียวกันด้วยการทำ standard scaler
scaler = StandardScaler()
X_sd = scaler.fit_transform(X)
````

````python
X_sd
````

````python
# Plot histograms for each scaled feature
for i in range(X_sd.shape[1]):
    plt.figure(figsize=(8, 6)) # Create a new figure for each plot
    plt.hist(X_sd[:, i], bins=12)
    plt.title(f'Histogram of {features[i]} (Scaled)')
    plt.xlabel(features[i])
    plt.ylabel('Frequency')
    plt.show() # Display each plot immediately
````

````python
#ปรับ scale ข้อมูลให้อยู่ในหน่วยเดียวกันด้วยการทำ min max scaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_mm = scaler.fit_transform(X)
````

#### Plot error to decisision K

###### StandardScaler

````python
# ทำการสร้างโมเดลและพล็อตค่า error ในแต่ละ k เพื่อดูว่า ควรเลือกใช้ K เป็นเท่าไหร่
#โดยดูที่ึการลดลงของ error ไม่ได้ลดลงอย่างค้มค่า หรือ ดูที่มุมหักศอก

inertia = []
K = range(1, 10)
for k in K:
    model = KMeans(n_clusters=k, random_state=42)
    model.fit(X_sd)
    inertia.append(model.inertia_)

plt.plot(K, inertia, 'o-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k with StandardScaler')
plt.show()
````

###### MinMaxScaler

````python
# ทำการสร้างโมเดลและพล็อตค่า error ในแต่ละ k เพื่อดูว่า ควรเลือกใช้ K เป็นเท่าไหร่
#โดยดูที่ึการลดลงของ error ไม่ได้ลดลงอย่างค้มค่า หรือ ดูที่มุมหักศอก

inertia = []
K = range(1, 10)
for k in K:
    model = KMeans(n_clusters=k, random_state=42)
    model.fit(X_mm)
    inertia.append(model.inertia_)

plt.plot(K, inertia, 'o-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k with MinMaxScaler')
plt.show()
````

````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import StandardScaler

# ข้อมูล
data = {
    'Cluster_sd': [0, 1, 2, 3],
    'CPI': [135.08, 215.11, 129.08, 135.08],
    'Unemployment': [7.77, 7.93, 13.74, 8.07],
    'Fuel_Price': [3.54, 3.29, 3.70, 3.69],
    'Sales_Per_Size': [0.08, 0.05, 0.05, 0.05]
}

cluster_profile = pd.DataFrame(data)

# เลือกคอลัมน์ที่จะทำ radar chart
categories = ['CPI', 'Unemployment', 'Fuel_Price', 'Sales_Per_Size']

# Standardize
scaler = StandardScaler()
cluster_std = cluster_profile.copy()
cluster_std[categories] = scaler.fit_transform(cluster_profile[categories])

# จำนวนแกน
N = len(categories)
angles = np.linspace(0, 2 * np.pi, N, endpoint=False).tolist()
angles += angles[:1]  # ปิดวง

# สร้าง radar chart
fig, ax = plt.subplots(figsize=(6,6), subplot_kw=dict(polar=True))

for i, row in cluster_std.iterrows():
    values = row[categories].tolist()
    values += values[:1]  # ปิดวง

    # Set line width based on whether any value is not equal to 0
    line_width = 2 if any(v != 0 for v in values[:-1]) else 1

    ax.plot(angles, values, label=f"Cluster {row['Cluster_sd']}", linewidth=line_width)
    ax.fill(angles, values, alpha=0.25)

# ปรับ labels
ax.set_xticks(angles[:-1])
# Add padding to the labels to move them away from the center
ax.set_xticklabels(categories, ha='center', va='center', fontsize=10, color='dimgray') # Adjust ha, va, fontsize, color as needed
ax.tick_params(axis='x', pad=20) # Add padding to the x-axis ticks


plt.title("Standardized Cluster Profile Radar Chart")
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1.1))
plt.show()
````

## 3. การสร้างตัวแบบทำนาย (Predictive Model Building)

เนื่องจากเป็นปัญหาการทำนายตัวแปรเชิงปริมาณ (Weekly_Sales) เราจะใช้ตัวแบบ Regression

### 3.1 การเตรียมข้อมูล (Prepare Data)


1. การรวมข้อมูล (Data Merging)

รวมตารางทั้งสี่เข้าด้วยกันโดยใช้คอลัมน์ Store และ Date เป็นกุญแจสำคัญ (Key).

2. การจัดการข้อมูลสูญหาย (Handle Missing Data)
- MarkDown1-5: ค่าว่างส่วนใหญ่น่าจะหมายถึง ไม่มีโปรโมชั่น ในสัปดาห์นั้น ดังนั้น เติมค่าว่างด้วย 0.

- CPI / Unemployment / Temperature / Fuel_Price: เติมค่าว่างที่เหลือด้วย ค่าเฉลี่ย (Mean) หรือ ค่าก่อนหน้า/ถัดไป (Forward/Backward Fill) ภายใน Store หรือ Date เดียวกัน.

- Weekly_Sales (ใน train.csv): ไม่มีค่าว่าง (ตามโครงสร้างข้อมูล) แต่ต้องจัดการ Outliers ที่สุดขั้ว (เช่น ยอดขายเป็นลบ) โดยอาจแทนที่ด้วยค่า 0 หรือกำจัดออก.

###3.2 วิศวกรรมคุณลักษณะ (Feature Engineering)

1. การแยกวันที่ (Date Features):

- สร้าง Year, Month, Week จากคอลัมน์ Date เพื่อจับแนวโน้มตามฤดูกาลและปี

- สร้าง DayOfWeek (เช่น 5=วันศุกร์)

2. คุณลักษณะรวม (Aggregate Features):

- MarkDown_Total: ผลรวมของ MarkDown1 ถึง MarkDown5 (เพื่อจับผลรวมของกิจกรรมส่งเสริมการขาย)

3. คุณลักษณะเชิงปฏิสัมพันธ์ (Interaction Features):

- Size * Type (การมีปฏิสัมพันธ์ระหว่างขนาดและประเภทของร้านค้า)

### 3.3 การแบ่งชุดข้อมูล (Partition Data)

- ตัวแปรเป้าหมาย (Y): Weekly_Sales

- ตัวแปรคุณลักษณะ (X): Store, Dept, IsHoliday, Type, Size, Year, Month, Week, Temperature, Fuel_Price, CPI, Unemployment, MarkDown_Total ฯลฯ

แบ่งข้อมูล train.csv ออกเป็น Training Set และ Validation Set (เช่น 80%/20%) โดยใช้ Time-Based Split เพื่อรักษาความต่อเนื่องของข้อมูลตามเวลา (Train บนข้อมูลเก่า, Validate บนข้อมูลใหม่)

- Handle Imbalance Data (N/A): เนื่องจากเป็นปัญหา Regression จึง ไม่มีปัญหาข้อมูลไม่สมดุล (Class Imbalance) เหมือนกับปัญหา Classification แต่เราจะจัดการกับ Outliers ใน Y แทน

### 3.4 การสร้างตัวแบบ (Build Model)
เลือกใช้ XGBoost Regressor (Extreme Gradient Boosting) ซึ่งเป็น Ensemble Method ที่มีประสิทธิภาพสูงสำหรับปัญหา Regression และจัดการกับข้อมูลที่มีคุณลักษณะและ Outliers ได้ดี

การเตรียมตัวแปร (Preprocessing Pipeline):

- Categorical Features (Store, Dept, Type, IsHoliday, Year, Month): ใช้ One-Hot Encoding

- Numerical Features (Size, Temperature, Fuel_Price, CPI, Unemployment, MarkDowns): ใช้ StandardScaler หรือ MinMaxScaler

- นำข้อมูลที่ผ่านการประมวลผลล่วงหน้าแล้วไปฝึกสอน XGBoost Regressor

### 3.5 การปรับ Hyperparameter (Hyperparameter Tuning)

ใช้ Grid Search หรือ Random Search ร่วมกับ Cross-Validation (Time-Series Cross-Validation) บน Training Set เพื่อหาค่าที่ดีที่สุดของ Hyperparameter สำคัญ:

- n_estimators (จำนวนต้นไม้)

- max_depth (ความลึกสูงสุดของต้นไม้)

- learning_rate (อัตราการเรียนรู้)

- subsample, colsample_bytree (อัตราส่วนของข้อมูลและคุณลักษณะที่ใช้ในแต่ละต้นไม้)

Metric สำคัญ: ใช้ Root Mean Squared Error (RMSE) เป็นตัววัดประสิทธิภาพหลักในการปรับจูน เนื่องจากเป็น Metric ที่ลงโทษความผิดพลาดที่สูงกว่า (Outliers) มากกว่า Mean Absolute Error (MAE)

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error
import xgboost as xgb

# กำหนดให้แสดงผลลัพธ์แบบเต็ม
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

# ----------------------------------------------------------------------
# 🌟 ส่วนที่ 0: การโหลดและรวมข้อมูล (Data Loading and Merging)
# ----------------------------------------------------------------------

print("--- 🌟 0. การโหลดและรวมข้อมูล ---")

# โหลดไฟล์ข้อมูล
try:
    df_train = pd.read_csv('/content/drive/MyDrive/Walmart/train.csv')
    df_test = pd.read_csv('/content/drive/MyDrive/Walmart/test.csv')
    df_features = pd.read_csv('/content/drive/MyDrive/Walmart/features.csv')
    df_stores = pd.read_csv('/content/drive/MyDrive/Walmart/stores.csv')
except FileNotFoundError as e:
    print(f"Error: {e}. Please ensure all four CSV files are in the same directory.")
    # หยุดการทำงานหากหาไฟล์ไม่พบ
    exit()

# แปลงคอลัมน์ 'Date' เป็นรูปแบบ datetime
df_train['Date'] = pd.to_datetime(df_train['Date'])
df_test['Date'] = pd.to_datetime(df_test['Date'])
df_features['Date'] = pd.to_datetime(df_features['Date'])

# รวมข้อมูล Store เข้ากับ Features
df_features = pd.merge(df_features, df_stores, on='Store', how='left')

# รวมข้อมูล Features เข้ากับ Training และ Testing set
train_df = pd.merge(df_train, df_features, on=['Store', 'Date', 'IsHoliday'], how='left')
test_df = pd.merge(df_test, df_features, on=['Store', 'Date', 'IsHoliday'], how='left')

print("ขนาดข้อมูลหลังรวม (Train):", train_df.shape)
print("ขนาดข้อมูลหลังรวม (Test):", test_df.shape)
print("\n")
````

````python
# ----------------------------------------------------------------------
# 📊 ส่วนที่ 1: การเล่าเรื่องราวจากข้อมูลและการสำรวจข้อมูล (Data Storytelling & EDA)
# ----------------------------------------------------------------------

print("--- 📊 1. การเล่าเรื่องราวและการสำรวจข้อมูลเบื้องต้น ---")

# 1.1 ลักษณะข้อมูล (Data Types and Missing Values)
print("ข้อมูลสูญหาย (Missing Values) ใน Training Set:")
print(train_df.isnull().sum()[train_df.isnull().sum() > 0])
print("---------------------------------------------")

# 1.2 สถิติเชิงพรรณนา (Descriptive Statistics)
print("สถิติ Weekly_Sales (ตัวแปรเป้าหมาย):")
print(train_df['Weekly_Sales'].describe())
print("---------------------------------------------")

# 1.3 การสำรวจข้อคาดการณ์ (Conjecture Analysis - ใช้กราฟ)
# ข้อคาดการณ์ที่ 1: ยอดขายวันหยุดสูงกว่า
sales_by_holiday = train_df.groupby('IsHoliday')['Weekly_Sales'].mean()
print("ค่าเฉลี่ย Weekly_Sales (IsHoliday=True/False):")
print(sales_by_holiday)
#

# ข้อคาดการณ์ที่ 3: ยอดขายตามประเภทร้าน
sales_by_type = train_df.groupby('Type')['Weekly_Sales'].mean().sort_values(ascending=False)
print("ค่าเฉลี่ย Weekly_Sales ตาม Type ของร้านค้า:")
print(sales_by_type)
#
print("\n")
````

````python
# ----------------------------------------------------------------------
# 🛠️ ส่วนที่ 2: การเตรียมข้อมูล (Data Preparation)
# ----------------------------------------------------------------------

print("--- 🛠️ 2. การเตรียมข้อมูลและวิศวกรรมคุณลักษณะ ---")

# 2.1 การจัดการข้อมูลสูญหาย (Missing Data Imputation)
# MarkDowns: เติมค่าว่างด้วย 0 (สมมติว่าไม่มีโปรโมชั่น)
markdown_cols = [f'MarkDown{i}' for i in range(1, 6)]
train_df[markdown_cols] = train_df[markdown_cols].fillna(0)
test_df[markdown_cols] = test_df[markdown_cols].fillna(0)

# CPI, Unemployment: เติมค่าว่างด้วยค่าเฉลี่ยของ Store นั้นๆ หรือใช้ค่าเฉลี่ยรวม
for col in ['CPI', 'Unemployment', 'Fuel_Price', 'Temperature']:
    train_df[col] = train_df[col].fillna(train_df.groupby('Store')[col].transform('mean'))
    test_df[col] = test_df[col].fillna(test_df.groupby('Store')[col].transform('mean'))
    # เติมค่าว่างที่เหลือด้วยค่าเฉลี่ยรวม (ถ้า Store นั้นไม่มีข้อมูล)
    train_df[col] = train_df[col].fillna(train_df[col].mean())
    test_df[col] = test_df[col].fillna(test_df[col].mean())


# 2.2 การจัดการ Outliers และยอดขายเป็นลบ
# กำจัดยอดขายที่เป็นลบออกจากชุดฝึกสอน (ไม่สามารถทำนายยอดขายติดลบในทางปฏิบัติ)
train_df = train_df[train_df['Weekly_Sales'] >= 0]


# 2.3 วิศวกรรมคุณลักษณะ (Feature Engineering)

def feature_engineer(df):
    # 1. คุณลักษณะจากวันที่
    df['Year'] = df['Date'].dt.year
    df['Month'] = df['Date'].dt.month
    df['Week'] = df['Date'].dt.isocalendar().week.astype(int)

    # 2. คุณลักษณะรวม (Total MarkDown)
    df['MarkDown_Total'] = df['MarkDown1'] + df['MarkDown2'] + df['MarkDown3'] + df['MarkDown4'] + df['MarkDown5']

    # 3. ตัวแปร Type เป็นตัวเลข
    df['Type_Encoded'] = df['Type'].map({'A': 3, 'B': 2, 'C': 1})

    # 4. คุณลักษณะปฏิสัมพันธ์ (Store size * Type)
    df['Size_Type_Interaction'] = df['Size'] * df['Type_Encoded']

    # 5. การเข้ารหัส IsHoliday เป็นตัวเลข
    df['IsHoliday'] = df['IsHoliday'].astype(int)

    return df

train_df = feature_engineer(train_df)
test_df = feature_engineer(test_df)

print("จำนวนคอลัมน์หลัง Feature Engineering:", train_df.shape[1])
print("\n")
````

````python
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from xgboost.callback import EarlyStopping # นำเข้า EarlyStopping

# NOTE: สมมติว่า train_df ถูกกำหนดไว้แล้วจากการรันโค้ดส่วนก่อนหน้า

print("--- 💾 3. การแบ่งชุดข้อมูลและการฝึกสอน ---")

# 3.1 การกำหนดตัวแปรเป้าหมายและคุณลักษณะ
FEATURES = [
    'Store', 'Dept', 'IsHoliday', 'Size', 'Temperature', 'Fuel_Price',
    'CPI', 'Unemployment', 'MarkDown1', 'MarkDown2', 'MarkDown3',
    'MarkDown4', 'MarkDown5', 'MarkDown_Total', 'Year', 'Month', 'Week',
    'Type_Encoded', 'Size_Type_Interaction'
]
TARGET = 'Weekly_Sales'

X = train_df[FEATURES]
y = train_df[TARGET]

# 3.2 Time-Based Split สำหรับ Training/Validation
# ใช้ข้อมูล 80% แรกสำหรับการฝึกสอน และ 20% หลังสำหรับการตรวจสอบ
split_index = int(len(X) * 0.8)

# แบ่งตามดัชนี (ลำดับเวลา)
X_train = X.iloc[:split_index]
X_val = X.iloc[split_index:]
y_train = y.iloc[:split_index]
y_val = y.iloc[split_index:] # ถูกต้องแล้ว: ใช้ 20% หลัง

print(f"Training Set Size: {len(X_train)} ({len(X_train)/len(X):.1%})")
print(f"Validation Set Size: {len(X_val)} ({len(X_val)/len(X):.1%})")

# 3.3 Pipeline สำหรับการประมวลผลล่วงหน้า (Preprocessing)
NUM_FEATURES = ['Size', 'Temperature', 'Fuel_Price', 'CPI', 'Unemployment',
                'MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4', 'MarkDown5',
                'MarkDown_Total', 'Size_Type_Interaction']

CAT_FEATURES = ['Store', 'Dept', 'IsHoliday', 'Year', 'Month', 'Week', 'Type_Encoded']

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), NUM_FEATURES),
        # FIX: ลบ handle_feature_names_out ออก (แก้ปัญหา Scikit-learn เวอร์ชันเก่า)
        ('cat', OneHotEncoder(handle_unknown='ignore'), CAT_FEATURES)
    ],
    remainder='passthrough'
)

# 3.4 การสร้างตัวแบบ (XGBoost Regressor)
xgb_model = xgb.XGBRegressor(
    objective='reg:squarederror',
    n_estimators=1000,
    learning_rate=0.05,
    max_depth=7,
    subsample=0.7,
    colsample_bytree=0.7,
    random_state=42,
    tree_method='hist',
    eval_metric='rmse', # กำหนด eval_metric ที่นี่
    n_jobs=-1
)

# 3.5 การสร้าง Pipeline และฝึกสอนตัวแบบ
# Fit preprocessor (แยกทำเพื่อให้ XGBoost ได้รับข้อมูลเมทริกซ์ที่พร้อมใช้งาน)
X_train_processed = preprocessor.fit_transform(X_train)
X_val_processed = preprocessor.transform(X_val)

print("กำลังฝึกสอนตัวแบบ XGBoost Regressor ด้วย Early Stopping...")

# 1. กำหนด Callback สำหรับ Early Stopping
early_stopping = EarlyStopping(
    rounds=50,
    metric_name='rmse',
    data_name='validation',
    # FIX: ลบ verbose ออก เพื่อแก้ปัญหา TypeError
)

# 2. เรียกใช้ .fit() พร้อมพารามิเตอร์ callbacks
xgb_model.fit(
    X_train_processed,
    y_train,
    eval_set=[(X_val_processed, y_val)],
    callbacks=[early_stopping]
)

print("ฝึกสอนเสร็จสิ้น.\n")
````
