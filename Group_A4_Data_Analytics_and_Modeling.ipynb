การพยากรณ์ยอดขายรายสัปดาห์ที่แม่นยำเป็นหัวใจสำคัญของการดำเนินงานร้านค้าปลีก:

- การบริหารสินค้าคงคลัง (Inventory Management): ช่วยลดต้นทุนการเก็บรักษาสินค้า (Holding Costs) และป้องกันการขาดแคลนสินค้า (Stockouts)

- การจัดสรรบุคลากร: วางแผนกำลังคนให้เพียงพอในช่วงยอดขายสูง

- การวางแผนกลยุทธ์การตลาด: เข้าใจว่าโปรโมชั่น (MarkDowns) ใดมีประสิทธิภาพสูงสุด

ข้อคาดการณ์ (Conjectures) 3 ข้อ

1. ข้อคาดการณ์ที่ 1 (วันหยุด): ยอดขายรายสัปดาห์ในช่วงสัปดาห์ที่เป็นวันหยุด (โดยเฉพาะวันหยุดใหญ่) จะมีค่าเฉลี่ย สูงกว่า สัปดาห์ปกติอย่างมีนัยสำคัญ

2. ข้อคาดการณ์ที่ 2 (CPI/Unemployment): ตัวชี้วัดทางเศรษฐกิจ CPI (เงินเฟ้อ) และ Unemployment (ว่างงาน) จะมีความสัมพันธ์ เชิงลบ กับยอดขาย (เมื่อเศรษฐกิจไม่ดี ยอดขายจะลดลง)

3. ข้อคาดการณ์ที่ 3 (MarkDowns): โปรโมชั่นลดราคา MarkDowns จะมีความสัมพันธ์ เชิงบวก กับยอดขาย แต่ MarkDown บางประเภทจะมีประสิทธิภาพสูงกว่าประเภทอื่น

  - การยืนยันว่า MarkDowns มีผล (ข้อ 3) จะนำไปสู่การ จัดสรรงบประมาณโปรโมชั่น ไปที่ MarkDown ประเภทที่มีประสิทธิภาพสูงสุด (MarkDown1/4) และวางแผนสินค้าที่จะลดราคาให้แม่นยำตามการพยากรณ์

## 3. การสร้างตัวแบบทำนาย (Predictive Model Building)

เนื่องจากเป็นปัญหาการทำนายตัวแปรเชิงปริมาณ (Weekly_Sales) เราจะใช้ตัวแบบ Regression

### 3.1 การเตรียมข้อมูล (Prepare Data)


1. การรวมข้อมูล (Data Merging)

รวมตารางทั้งสี่เข้าด้วยกันโดยใช้คอลัมน์ Store และ Date เป็นกุญแจสำคัญ (Key).

2. การจัดการข้อมูลสูญหาย (Handle Missing Data)
- MarkDown1-5: ค่าว่างส่วนใหญ่น่าจะหมายถึง ไม่มีโปรโมชั่น ในสัปดาห์นั้น ดังนั้น เติมค่าว่างด้วย 0.

- CPI / Unemployment / Temperature / Fuel_Price: เติมค่าว่างที่เหลือด้วย ค่าเฉลี่ย (Mean) หรือ ค่าก่อนหน้า/ถัดไป (Forward/Backward Fill) ภายใน Store หรือ Date เดียวกัน.

- Weekly_Sales (ใน train.csv): ไม่มีค่าว่าง (ตามโครงสร้างข้อมูล) แต่ต้องจัดการ Outliers ที่สุดขั้ว (เช่น ยอดขายเป็นลบ) โดยอาจแทนที่ด้วยค่า 0 หรือกำจัดออก.

###3.2 วิศวกรรมคุณลักษณะ (Feature Engineering)

1. การแยกวันที่ (Date Features):

- สร้าง Year, Month, Week จากคอลัมน์ Date เพื่อจับแนวโน้มตามฤดูกาลและปี

- สร้าง DayOfWeek (เช่น 5=วันศุกร์)

2. คุณลักษณะรวม (Aggregate Features):

- MarkDown_Total: ผลรวมของ MarkDown1 ถึง MarkDown5 (เพื่อจับผลรวมของกิจกรรมส่งเสริมการขาย)

3. คุณลักษณะเชิงปฏิสัมพันธ์ (Interaction Features):

- Size * Type (การมีปฏิสัมพันธ์ระหว่างขนาดและประเภทของร้านค้า)

### 3.3 การแบ่งชุดข้อมูล (Partition Data)

- ตัวแปรเป้าหมาย (Y): Weekly_Sales

- ตัวแปรคุณลักษณะ (X): Store, Dept, IsHoliday, Type, Size, Year, Month, Week, Temperature, Fuel_Price, CPI, Unemployment, MarkDown_Total ฯลฯ

แบ่งข้อมูล train.csv ออกเป็น Training Set และ Validation Set (เช่น 80%/20%) โดยใช้ Time-Based Split เพื่อรักษาความต่อเนื่องของข้อมูลตามเวลา (Train บนข้อมูลเก่า, Validate บนข้อมูลใหม่)

- Handle Imbalance Data (N/A): เนื่องจากเป็นปัญหา Regression จึง ไม่มีปัญหาข้อมูลไม่สมดุล (Class Imbalance) เหมือนกับปัญหา Classification แต่เราจะจัดการกับ Outliers ใน Y แทน

### 3.4 การสร้างตัวแบบ (Build Model)
เลือกใช้ XGBoost Regressor (Extreme Gradient Boosting) ซึ่งเป็น Ensemble Method ที่มีประสิทธิภาพสูงสำหรับปัญหา Regression และจัดการกับข้อมูลที่มีคุณลักษณะและ Outliers ได้ดี

การเตรียมตัวแปร (Preprocessing Pipeline):

- Categorical Features (Store, Dept, Type, IsHoliday, Year, Month): ใช้ One-Hot Encoding

- Numerical Features (Size, Temperature, Fuel_Price, CPI, Unemployment, MarkDowns): ใช้ StandardScaler หรือ MinMaxScaler

- นำข้อมูลที่ผ่านการประมวลผลล่วงหน้าแล้วไปฝึกสอน XGBoost Regressor

### 3.5 การปรับ Hyperparameter (Hyperparameter Tuning)

ใช้ Grid Search หรือ Random Search ร่วมกับ Cross-Validation (Time-Series Cross-Validation) บน Training Set เพื่อหาค่าที่ดีที่สุดของ Hyperparameter สำคัญ:

- n_estimators (จำนวนต้นไม้)

- max_depth (ความลึกสูงสุดของต้นไม้)

- learning_rate (อัตราการเรียนรู้)

- subsample, colsample_bytree (อัตราส่วนของข้อมูลและคุณลักษณะที่ใช้ในแต่ละต้นไม้)

Metric สำคัญ: ใช้ Root Mean Squared Error (RMSE) เป็นตัววัดประสิทธิภาพหลักในการปรับจูน เนื่องจากเป็น Metric ที่ลงโทษความผิดพลาดที่สูงกว่า (Outliers) มากกว่า Mean Absolute Error (MAE)
